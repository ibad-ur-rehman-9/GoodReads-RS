```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
data_path = "../data/raw/interactions_large.csv"
df = pd.read_csv(data_path)

df.head()
```

```python
ratings_df = df[['user_id', 'book_id', 'rating']]
ratings_df.head()
```

```python
ratings_df = ratings_df[ratings_df['rating'] > 0]
ratings_df.shape
```

```python
ratings_df['rating'].value_counts().sort_index()
```

```python
num_users = ratings_df['user_id'].nunique()
num_books = ratings_df['book_id'].nunique()
num_ratings = ratings_df.shape[0]

num_users, num_books, num_ratings
```

```python
user_counts = ratings_df['user_id'].value_counts()
book_counts = ratings_df['book_id'].value_counts()

ratings_df = ratings_df[
    ratings_df['user_id'].isin(user_counts[user_counts >= 5].index) &
    ratings_df['book_id'].isin(book_counts[book_counts >= 5].index)
]

ratings_df.shape
```

```python
output_path = "../data/processed/ratings_clean.csv"
ratings_df.to_csv(output_path, index=False)
```
